Import all the necessary packages. Using preexisting packages from scikit-learn for data splitting and evaluation

```python
#Import Packages
#data management
import pandas as pd

#for training
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# for plotting
import matplotlib.pyplot as plt 
```

Read data from the Salary_Data_sample.csv file using panda

```python
#read data from excel sheey
data = pd.read_csv('Salary_Data_sample.csv')
print(data.head())
```

Finding data characteristics and data splitting. The data is random divied when splitting

```python
# Find characteristics of 
data.info()

# Split data RANDOMLY such test data comprises 1/3 of total data
train_data, test_data = train_test_split(data, test_size=1/3)
```

Use a simple Linear Regressor for modelling the data as it is a relatively simple dataset with only 2 columns and 30 entries

```python
# create a linear regression model
model = LinearRegression()

# fit the model to the training data
model.fit(train_data[['YearsExperience']], train_data['Salary'])
```

Predict what the output (Salary) would be according to the trained models for the data points under YearsExperience. Also find the mean squared error.

```python
# Predicting the Test set result ￼
Y_Pred = model.predict(test_data[['YearsExperience']])
```

```python
# calculate the mean squared error
mse = mean_squared_error(test_data['Salary'], Y_Pred)

# print the mean squared error
print(f"Mean squared error: {mse}")
```

Generate a scatterplot to show the testing and training data distribution and best fit line accoring of the trained Linear Regressor model

```python
# create a scatter plot of the training data
plt.scatter(train_data['YearsExperience'], train_data['Salary'], label='Training data')

# create a scatter plot of the test data
plt.scatter(test_data['YearsExperience'], test_data['Salary'], label='Test data')

# add a line of best fit to the plot
plt.plot(train_data['YearsExperience'], model.predict(train_data[['YearsExperience']]), color='red', label='Best fit line')

# add labels and a legend to the plot
plt.xlabel('Duration')
plt.ylabel('Salary')
plt.legend()

# show the plot
plt.show()
```
